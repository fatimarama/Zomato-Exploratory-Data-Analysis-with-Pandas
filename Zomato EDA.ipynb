```python
# original data frame is not changed here

df3 = df["Cuisines"].fillna("Not Availabale")
df3
```

```python
# change actual data set
df["Cuisines"].fillna("Not Available", inplace = True)
df
```

```python
df_Cuisines = df[df["Cuisines"] == "Not Available"]
df_Cuisines
```

```python
# drops rows with missing values
df.dropna(inplace = True)
```

```python
# axis = 0 means row, axis = 1 is column
# drop a column

df.drop('Latitude', axis=1, inplace=True)
df
```

```python
# drop duplicates
# how to check drop values
duplicates = df.drop_duplicates(inplace = True)
duplicates
```

```python
#drop duplicates based on a specific value in a column

df = df[~(df['Cuisines'] == 'Not Available')].drop_duplicates()
```

```python
#adding a new column in the df

df['High Rated'] = df['Aggregate rating'] / 2
df['High Rated']
```

```python
#using lambda function 

df['High Rated'] = df['Aggregate rating'].apply(lambda x: 'Yes' if x > 4.0 else 'No')
df['High Rated']
```

```python
# mean of all column

df['Aggregate rating'].mean()
```

```python
# mean according to the city column

df.groupby("City")['Aggregate rating'].mean()
```
